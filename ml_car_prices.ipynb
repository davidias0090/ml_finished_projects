#Bibliotecas e Leitura dos Dados



```python
import pandas as pd
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
# Construção do modelo machine learning
# Treino 75% dados - fit
# Teste 25% - predict -> Acuracia(verificao se a bom ou ruim o modelo que foi treinado)
# Se ACURACIA boa JOGA O MODELO NO MUNDO!!!
url = 'https://raw.githubusercontent.com/davidias0090/ml_finished_projects/main/car-prices.csv' # Use raw file URL for direct download
df = pd.read_csv(url)
df.head(10)
```

## Renomeando Colunas e separando features

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np


mapa = {'mileage_per_year' : 'milhas_por_ano', 'model_year' : 'ano_modelo', 'price' : 'preco', 'sold' : 'vendido'}
df = df.rename(columns=mapa)

```

```python
#Trocando milhas por ano por Kilometros por ano
df['milhas_por_ano'] = df.milhas_por_ano * 1.60934
#Trocando nome da coluna milhas_por_ano por KM_por_ano
df = df.rename(columns={'milhas_por_ano' : 'km_por_ano'}) # This line seems redundant now
#Trocar os valores yes e no da coluna vendido para 1 e 0
troca = {
    "yes": 1,
    'no': 0
}
df['vendido'] = df['vendido'].map(troca)
df.head(10)
```

```python
x = df[['preco', 'ano_modelo', 'km_por_ano']]
y = df['vendido']
```

#Separação do Treino e Teste

```python
seed = 20
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.25, random_state=seed)

print('Treinamento com %d elementos e testarmos com %d elementos' % (len(train_x), len(test_x)))
```

## Modelo Classificador

```python
model = LinearSVC() # Cria o modelo
model.fit(train_x, train_y) # treino

result = model.predict(test_x) # teste

acuracia = accuracy_score(test_y, result) # compara vetores
acuracia_model = model.score(test_x, test_y) # roda teste

print("Acuracia_model: ", (acuracia_model*100), "%")
```

## COMPARAÇÃO ACURACIA DUMMY (BASELINE)

```python
from sklearn.dummy import DummyClassifier

dummy_stratified = DummyClassifier()
dummy_stratified.fit(train_x, train_y)
acuracia_dummy = dummy_stratified.score(test_x, test_y)

print("Acuracia Dummy: ", (acuracia_dummy*100), "%")
```
